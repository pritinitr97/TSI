```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Summer

```python
df=pd.read_excel("C:\\Users\\priti\\Documents\\chlorophyll.xlsx")
```

```python
csfont = {'fontname':'Times New Roman'}
plt.figure(figsize=(4,4), tight_layout=True)
tsi1=[]
for (p,v) in df.iterrows():
    k=0
    chl=0
    
    for i in v:
        if k==1:
            chl=i
        k=k+1
    tsi=9.81* np.log(chl)+30.60
    print("TSI value :",round(tsi,2))
    tsi1.append(round(tsi,2))
print(tsi1)
plt.plot(tsi1,'o-r')
plt.title(' Summer \n Trophic State Index ',weight='bold',**csfont)
plt.xlabel('Samples',weight='bold',**csfont)
plt.ylabel('TSI',weight='bold',**csfont)
```

```python
csfont = {'fontname':'Times New Roman'}
plt.figure(figsize=(4,4), tight_layout=True)
#plt.plot(sample,chlo,'o-r')
#plt.plot(sample,sr,'s--g')
x=df['Sample No.']
y=df['Chlorophyll (μg/l)']
plt.plot(x,y,'o-r')
plt.plot(x,tsi1,'s--g')
plt.title(' Summer \n Variation of TSI w.r.t Chlorophyll',weight='bold',**csfont)
plt.xlabel('Samples',weight='bold',**csfont)
plt.ylabel('TSI Range',weight='bold',**csfont)
plt.legend(['Chlorophyll (µg/l)','TSI'])
plt.xticks(fontweight = 'bold',**csfont)
plt.yticks(fontweight = 'bold',**csfont)
plt.ylim(0, 40)
plt.xlim(0, 20)
plt.grid(True, ls="dotted")
plt.rcParams["font.weight"] = "bold"
plt.rcParams["font.serif"] = ["Times New Roman"]
```

```python
ma_tsi=np.array(tsi1)
m=ma_tsi.mean()
print(round(m,2))
```
